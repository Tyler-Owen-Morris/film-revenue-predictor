# Looking to source title codes for the data-set I'm going to scrape

```python
titles_df = pd.read_csv('data/title.basics.tsv', sep='\t')
aka_df = pd.read_csv('data/title.akas.tsv', sep='\t')
```

```python
aka_df.head()
```

```python
aka_df.region.unique()
```

```python
titles2015_df = titles_df[titles_df['startYear']=='2014']
titles2015_df = titles2015_df[titles2015_df['titleType']=='movie']
titles2015_df = titles2015_df[titles2015_df['isAdult']==0]
titles2015_df.info()
```

```python
titles2015_df.head(10)
```

```python
us_df = aka_df[aka_df['region']=='US']
us_df.drop_duplicates(subset = 'titleId', keep='first', inplace=True)
us_df.head()
```

```python
titles2015_df.head()
```

```python
titles2015_df.drop_duplicates(subset = 'tconst', keep='first', inplace=True)
titles2015_df = titles2015_df.rename(columns={"tconst":"titleId"})
```

```python
titles2015_df.head()
```

```python
us2015movie_df = titles2015_df.merge(us_df, on='titleId', how='inner')
us2015movie_df.head()
```

```python
us2015movie_df[['titleId', 'primaryTitle']].head()
```

```python
us2015movie_df[['titleId', 'primaryTitle']].to_csv('data/us2014_filmList.csv')
```
